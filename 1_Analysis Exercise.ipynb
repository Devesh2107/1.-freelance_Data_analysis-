```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("1_Analysis Exercise_Updated.xlsx", sheet_name='Interview Exercise', header=4)
```

```python
df
```

```python
df.columns
```

```python
df = df.dropna()
```

```python
df
```

```python
df.describe()
```

```python
sales_by_region = df.groupby('Region').agg({'Units': 'sum', 'Unit Cost ($)': 'mean'})
```

```python
sales_by_region
```

```python
sales_by_item = df.groupby('Item').agg({'Units': 'sum', 'Unit Cost ($)': 'mean'})
```

```python
sales_by_item
```

```python
df.loc[:, 'Date'] = pd.to_datetime(df['Date'])
sales_over_time = df.groupby(df['Date'].dt.to_period('M')).agg({'Units': 'sum'})
```

```python
sales_over_time
```

```python
df.loc[:, 'Date']
```

```python
sales_over_time.plot(kind='line')
plt.title('Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend()
plt.show()
```

```python
sales_over_time.plot(kind='bar')
plt.title('Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend()
plt.show()
```

```python
df["Region"].value_counts().plot.pie(y = df["Region"], figsize = (12, 12), autopct = '%1.1f%%')
```

```python
df["Item"].value_counts().plot.pie(y = df["Item"], figsize = (8, 8), autopct = '%1.1f%%')
plt.legend()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

sales_trend = df.groupby('Date')['Units'].sum()

plt.figure(figsize=(10, 6))
plt.plot(sales_trend, marker='o', color='b')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Units')
plt.grid(True)
plt.show()
```
